### a) 1. Write your own plain gradient decent with a tuned learning rate
The goal of the minimization is to find a minima in the costfunction. In GD one always moves in the direction of the negative gradient $-\nabla F(\mathbf{x})$, so that 
$$\mathbf{x}_{k+1} = \mathbf{x}_k - \gamma_k \nabla F(\mathbf{x}_k) $$
where $\gamma_k$ is the learning rate or step length. 
However, this method risks being stuck in a local minima. 

#### Tuning the learning rate methods:
* Teste ulike verdier
* Bruke egenverdiene av Hessian matrisen - forklare dette!! (Forutsetter at matrisen er inverterbar)
* learning-schedule(t) = t/t-t0: Jo flere tidssteg, jo mindre blir learning raten. Dette er en dårlig metode...
* Adagrad: $x_{t+1} = x_t - \eta G^{-1/2}g_t$, hvor $g_t$ er gradienten, og $G^{-1/2}= \Sigma g_t g_t^{T}$. Hvor $\eta G^{-1/2}$ er learning raten. Det vil si at når det er lite endring i gradienten, blir $G$ stor, altså learning raten liten. 

```python
def OLS(X, y): 
    return npy.linalg.pinv(X.T @ X) @ X.T @ y
```

```python
def CostOLS(beta):
    return (1.0/n)*npy.sum((y-X @ beta)**2)
```

```python
def ridge(X, y, hyp): 
    n = (X.T @ X).shape[0] # get an identity matrix of the approproate size
    return npy.linalg.pinv((X.T @ X)-(hyp * npy.identity(n))) @ X.T @ y
```

```python
def MSE(y_data,y_model):
    n = npy.size(y_model)
    return npy.sum((y_data-y_model)**2)/n
```

```python
def gradient_descent(X, y, method, lmbda, stopping_criterion, solver='analytical', learning_rate=None, momentum=None): 
    
    #print(f"Learning rate {learning_rate} with momentum {momentum}")
    
    if not learning_rate:
        # use the eigenvalues of the Hessian matrix to tune the learning rate
        H = (2.0/n)* X.T @ X
        # Get the eigenvalues
        EigValues, EigVectors = npy.linalg.eig(H)
        step_size = 1.0/npy.max(EigValues) 
    else:
        step_size = learning_rate
    
    # For GD the smartest way to choose starting point is randomly, unless you can expect something from the beta values
    beta = npy.random.randn(degree+1,1) 
    change = 0.0
    stop = False
    
    solutions, scores = [], []
    
    while stop==False:
        # find gradient using the Hessian, from deriving the cost-function
        if method == "OLS":
            
            if solver == "analytical": 
                gradient = (2.0/n)*X.T @ (X @ beta-y) 
            elif solver=="autograd": 
                training_gradient = grad(CostOLS)
                gradient = training_gradient(beta)
            elif solver=="jax": #currently not working
                grad = jax_grad(CostOLS)
                beta = jnp.float64(beta)
                step_size = jnp.float64(learning_rate)
                gradient = grad(beta)
            else: print("Solver not defined")
                
        elif method == "Ridge":
            gradient = (2.0/n)*X.T @ (X @ (beta)-y)+2*lmbda*beta
            
        else: print("Method is undefined.")
            
        if npy.linalg.norm(gradient) < stopping_criterion: stop=True
        if momentum: 
            # part 2: with momentum
            new_change = step_size*gradient + momentum*change
            beta  = beta - new_change
            change = new_change
        else:     
            # part 1: without momentum update beta based on the learning rate * gradient
            beta  = beta - step_size*gradient
        #solution_eval = objective(beta)
        solution_eval = 0 # for testing
        
        solutions.append(beta)
        scores.append(solution_eval)
        
    return [solutions, scores]
```

### a) 3. Repeat these steps for stochastic gradient descent with mini batches and a given number of epochs. 
Use a tunable learning rate as discussed in the lectures from week 39. Discuss the results as functions of the various parameters (size of batches, number of epochs etc)

For SGD is a good idea to use a learning scheduler, rather than tuning using a grid search. For GD this method is not recommended because it will result in finding local minima, but for mant batches this shouldn't matter. 

```python
def learning_schedule(t, t0, t1):
    return t0/(t+t1)
```

```python
# TODO in this function: implement jax, get autograd working, implement autograd and jax for ridge
# TODO get this working as a GD algorithm 
# Make it a class
def SGD(X, y, n_epochs, batch_size, cost_function="OLS", solver="analytical", tuning_method="SetByUser", momentum=None, learning_rate=0.1): 
    
    n = X_train.shape[0]
    m = int(n/batch_size) #number of minibatches
    t0, t1 = 5, 50
    
    # intitalizations
    beta = npy.random.randn(degree+1,1)
    change = 0.0
    eta = 0.01
    delta  = 1e-8 # Including AdaGrad parameter to avoid possible division by zero
    rho = 0.99 # for RMSprop, controls the length scale of the moving average.
    beta1 = 0.9 # for ADAM, value see https://arxiv.org/abs/1412.6980
    beta2 = 0.999 # for ADAM, value see https://arxiv.org/abs/1412.6980
    iter = 0
    
    solutions = []
    
    # for autograd
    if cost_function=="OLS": training_gradient = grad(CostOLS)
    
    for epoch in range(n_epochs):
        
        Giter = 0.0 #AdaGrad and RMSprop
        first_moment = 0.0 # ADAM
        second_moment = 0.0 # ADAM
        iter += 1 # ADAM
        
        for i in range(m):
            random_index = batch_size*npy.random.randint(m)
            xi = X[random_index:random_index+batch_size]
            yi = y[random_index:random_index+batch_size]
            
            # Choice of cost functions
            if cost_function=="OLS":
                
                if solver=="analytical":
                    gradient = 2.0/n*xi.T @ ((xi @ beta)-yi)
                elif solver=="autograd":
                    gradient = training_gradient(beta)
                elif solver=="jax": continue # TODO: to be implemented
                else: print("Solver not defined")
                    
            elif cost_function=="Ridge": # TODO: implement more solvers for Ridge
                gradient = (2.0/n)*xi.T @ (xi @ (beta)-yi)+2*lmbda*beta
            else: print("Cost function is undefined")
                
            # tune the learning rate
            if tuning_method == "SetByUser": 
                eta = learning_rate
                v = eta*gradient
            elif tuning_method == "Hessian":
                H = (2.0/n)* xi.T @ xi
                EigValues, EigVectors = npy.linalg.eig(H)
                eta = 1.0/npy.max(EigValues) 
                v = eta*gradient
            elif tuning_method == "learning_schedule": 
                eta = learning_schedule(epoch*m+i, t0, t1)
                v = gradient*eta
            elif tuning_method == "AdaGrad":
                # scales the learning rate inversly by the square root of all the historical squared values of the gradient
                # -> greater progress in the more gently sloped directions of parameter space
                Giter += gradient*gradient 
                v = gradient*eta/(delta+np.sqrt(Giter))
            elif tuning_method == "RMSprop":
                Giter = (rho*Giter+(1-rho)*gradient*gradient) # Scaling with rho the new and the previous results
                v = gradient*eta/(delta+np.sqrt(Giter))
            elif tuning_method == "ADAM":
                first_moment = beta1*first_moment + (1-beta1)*gradient
                second_moment = beta2*second_moment+(1-beta2)*gradient*gradient
                first_term = first_moment/(1.0-beta1**iter)
                second_term = second_moment/(1.0-beta2**iter)
                v = eta*first_term/(np.sqrt(second_term)+delta)                
                
            # With and without momentum
            if momentum: 
                # part 2: with momentum
                new_change = v + momentum*change
                beta  = beta - new_change
                change = new_change
            else:     
                # part 1: without momentum update beta based on the learning rate * gradient
                beta  = beta - v
        solutions.append(beta)
            
    return (solutions)
```

To measure the performance of our classification problem we use the
so-called *accuracy* score.  The accuracy is as you would expect just
the number of correctly guessed targets $t_i$ divided by the total
number of targets, that is

$$
\text{Accuracy} = \frac{\sum_{i=1}^n I(t_i = y_i)}{n} ,
$$

where $I$ is the indicator function, $1$ if $t_i = y_i$ and $0$
otherwise if we have a binary classification problem. Here $t_i$
represents the target and $y_i$ the outputs of your FFNN code and $n$ is simply the number of targets $t_i$.
