```python
def OLS(X, y): 
    return npy.linalg.pinv(X.T @ X) @ X.T @ y
```

```python
def CostOLS(beta):
    return (1.0/n)*npy.sum((y-X @ beta)**2)
```

```python
def ridge(X, y, hyp): 
    n = (X.T @ X).shape[0] # get an identity matrix of the approproate size
    return npy.linalg.pinv((X.T @ X)-(hyp * npy.identity(n))) @ X.T @ y
```

```python
def MSE(y_data,y_model):
    n = npy.size(y_model)
    return npy.sum((y_data-y_model)**2)/n
```

```python
# TODO in this function: implement jax, get autograd working, implement autograd and jax for ridge
# TODO get this working as a GD algorithm 
# Make it a class
def SGD(X, y, n_epochs, batch_size, cost_function="OLS", solver="analytical", tuning_method="SetByUser", momentum=None, learning_rate=0.1): 
    
    n = X_train.shape[0]
    m = int(n/batch_size) #number of minibatches
    t0, t1 = 5, 50
    
    # intitalizations
    beta = npy.random.randn(degree+1,1)
    change = 0.0
    eta = 0.01
    delta  = 1e-8 # Including AdaGrad parameter to avoid possible division by zero
    rho = 0.99 # for RMSprop, controls the length scale of the moving average.
    beta1 = 0.9 # for ADAM, value see https://arxiv.org/abs/1412.6980
    beta2 = 0.999 # for ADAM, value see https://arxiv.org/abs/1412.6980
    iter = 0
    
    solutions = []
    
    # for autograd
    if cost_function=="OLS": training_gradient = grad(CostOLS)
    
    for epoch in range(n_epochs):
        
        Giter = 0.0 #AdaGrad and RMSprop
        first_moment = 0.0 # ADAM
        second_moment = 0.0 # ADAM
        iter += 1 # ADAM
        
        for i in range(m):
            random_index = batch_size*npy.random.randint(m)
            xi = X[random_index:random_index+batch_size]
            yi = y[random_index:random_index+batch_size]
            
            # Choice of cost functions
            if cost_function=="OLS":
                
                if solver=="analytical":
                    gradient = 2.0/n*xi.T @ ((xi @ beta)-yi)
                elif solver=="autograd":
                    gradient = training_gradient(beta)
                elif solver=="jax": continue # TODO: to be implemented
                else: print("Solver not defined")
                    
            elif cost_function=="Ridge": # TODO: implement more solvers for Ridge
                gradient = (2.0/n)*xi.T @ (xi @ (beta)-yi)+2*lmbda*beta
            else: print("Cost function is undefined")
                
            # tune the learning rate
            if tuning_method == "SetByUser": 
                eta = learning_rate
                v = eta*gradient
            elif tuning_method == "Hessian":
                H = (2.0/n)* xi.T @ xi
                EigValues, EigVectors = npy.linalg.eig(H)
                eta = 1.0/npy.max(EigValues) 
                v = eta*gradient
            elif tuning_method == "learning_schedule": 
                eta = learning_schedule(epoch*m+i, t0, t1)
                v = gradient*eta
            elif tuning_method == "AdaGrad":
                # scales the learning rate inversly by the square root of all the historical squared values of the gradient
                # -> greater progress in the more gently sloped directions of parameter space
                Giter += gradient*gradient 
                v = gradient*eta/(delta+np.sqrt(Giter))
            elif tuning_method == "RMSprop":
                Giter = (rho*Giter+(1-rho)*gradient*gradient) # Scaling with rho the new and the previous results
                v = gradient*eta/(delta+np.sqrt(Giter))
            elif tuning_method == "ADAM":
                first_moment = beta1*first_moment + (1-beta1)*gradient
                second_moment = beta2*second_moment+(1-beta2)*gradient*gradient
                first_term = first_moment/(1.0-beta1**iter)
                second_term = second_moment/(1.0-beta2**iter)
                v = eta*first_term/(np.sqrt(second_term)+delta)                
                
            # With and without momentum
            if momentum: 
                # part 2: with momentum
                new_change = v + momentum*change
                beta  = beta - new_change
                change = new_change
            else:     
                # part 1: without momentum update beta based on the learning rate * gradient
                beta  = beta - v
        solutions.append(beta)
            
    return (solutions)
```

the number of correctly guessed targets $t_i$ divided by the total
number of targets, that is

$$
\text{Accuracy} = \frac{\sum_{i=1}^n I(t_i = y_i)}{n} ,
$$

where $I$ is the indicator function, $1$ if $t_i = y_i$ and $0$
otherwise if we have a binary classification problem. Here $t_i$
represents the target and $y_i$ the outputs of your FFNN code and $n$ is simply the number of targets $t_i$.
